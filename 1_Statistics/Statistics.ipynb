### 2.3 Quartiles

1. Quantiles break data into equal parts.
2. Quartiles are a type of Quantiles that break data into 4 equal parts.

```python
def count_quartiles(arr):
    """
    Function to count elements in each quartile.
    """
    print("Total length:", arr.size)

    mask = arr <= np.percentile(arr, 25)
    q1 = arr[mask]
    print("Q1:", q1.size)

    mask = (arr > np.percentile(arr, 25)) & (arr <= np.percentile(arr, 50))
    q2 = arr[mask]
    print("Q2:", q2.size)

    mask = (arr > np.percentile(arr, 50)) & (arr <= np.percentile(arr, 75))
    q3 = arr[mask]
    print("Q3:", q3.size)

    mask = arr > np.percentile(arr, 75)
    q4 = arr[mask]
    print("Q4:", q4.size)
```

```python
count_quartiles(df["Height"])
```

```python
count_quartiles(df["Weight"])
```
