## 1 Hypothesis Testing

1. Hypothesis Testing is a branch of Inferential Statistics that helps us in breaking down the data and take trustable decisions.
2. For the given data, values that we see in Descriptive Statistics are statistically significant or not?
3. Hypothesis Testing helps is quantifying the statistical significance of values seen (as part Descriptive Statistics) in data.
4. Hypothesis Testing is not about predictions.

Hypothesis testing:


Hypothesis -> Experiment (controlled) -> Handful (Sampling) -> Unbiased -> Experiment output -> Validation

### 1.1 Basic terminologies

Process of designing some experiments, collecting data/evidence to prove a case in point i.e., measuring the impact of the experiment.

There can be three outcomes of impact:

1. No Impact
2. Positive Impact
3. Negative Impact

Hypothesis testing tries to validate the significance of claimed positive or negative impact over default current state of no impact.

#### Key Concepts

1. Assumptions
2. Data / Evidence to support or negate the assumption.

#### Aliases

1. Default Assumptions
2. Claim

> **Note**:
> 
> Burden of proof is on those who want to reject default assumption.

When there is enough data or evidence to backup the claim then the default assumption is rejected.

### 1.2 Null vs Alternate Hypothesis

#### Definition

1. Default assumption is called as $H_0$ - Null Hypothesis
2. Opposite of Default assumption is called as $H_a$ - Alternate Hypothesis

Other terms used for Default Assumptions:

1. Null Hypothesis $H_0$
2. Status Quo
3. Baseline Assumptions

Hypothesis Testing will have two outcomes:

1. Successfully reject $H_0$ or accept $H_a$
2. Successfully reject $H_a$ or failed to reject $H_0$

> **Note**:
>
> 1. There is no term or process to **accept Null Hypothesis**.
> 2. We gather evidence towards our claim to either **reject Null Hypothesis** or **failed to reject Null Hypothesis**.
> 3. We gather evidence towards our claim to either **accept Alternate Hypothesis** or **reject Alternate Hypothesis**.

### 1.3 p-value

#### Definition

$\text{p-value}$ is probability of observing data as extreme as what was observed given that Null Hypothesis $H_0$ is true.

$\text{p-value}$ is probability of all events which are more extreme than the given event that we are trying to analyze.

We reject $H_0$ when $\text{p-value}$ is very low.

> **Note**:
> 
> $\text{p-value}$ is not probability of $H_0$ or $H_a$.

#### Formula

Mathematical representation of $\text{p-value}$:  

$\text{p-value} = P(data \mid \text{$H_0$ is True})$

> **Note**:
> 
> Smaller the $\text{p-value}$ higher is the confidence in our claim.

### 1.4 Significance level and Confidence level

Hypothesis testing builds on Descriptive Statistics by assessing whether observed sample summaries (means, proportions, correlations, etc.)  
are statistically significant deviations from values implied by a null hypothesis, under an assumed probabilistic model.

#### Definition

$
\begin{array}{l}
\text{if p-value} \le \alpha \text{:} \\[2pt]
\quad \text{Reject $H_0$ i.e., Accept $H_a$} \\[8pt]
\text{else:} \\[2pt]
\quad \text{Failed to reject $H_0$ i.e., Reject $H_a$}
\end{array}
$

1. $\alpha$ is called as **Significance level**
2. $1 - \alpha$ is called as **Confidence level**

> **Note**:
>
> Generally in most of the industries 95% confidence level is considered hence  
> most common value for Confidence Level is 0.95 and Significance Level is 0.05

#### Threshold Value

1. Threshold value of $\alpha$ depends on the domain of the problem.
2. For example medical domain can have very low threshold of 0.01 or 1%.
3. Generally the value of $\alpha$ must be less than 0.05 or 5%.

#### Risks involved in Significance level

Setting a Significance level while taking a decision often involves some risk (or results in errors).  
There can be errors when calculating the $\text{p-value}$, which affects in the decision making.

```
                      Actual

             +----------+----------+

```

1. Type 1 Error $(\alpha)$ - False Positive - Significance level
2. Type 2 Error $(\beta)$ - False Negative

#### Example

Given below,  

$H_0 : \text{Person is Innocent}$  
$H_a : \text{Person is Guilty}$  

```
                                       Actually             
                            Innocent(-)        Guilty(+)     
                        +----------------+----------------+
            Innocent(-) | True Negative  | False Negative |
Judge said              +----------------+----------------+
             Guilty(+)  | False Positive | True Positive  |
                        +----------------+----------------+
```

1. Type-1 Error occurs in above example om rejecting $H_0$ when it is actually True.
2. Type-2 Error occurs in above example on failing to reject $H_0$ when it is actually False.

or

<div style="display: inline-block">

| Error                   | Description                               |
| :---------------------- | :---------------------------------------- |
| Type-1 (False Positive) | Rejecting $H_0$ when it is actually True. |
| Type-2 (False Negative) | Rejecting $H_a$ when it is actually True. |

</div>

> **Note**:
>
> 1. $H_0$ is the status quo and is often considered Negative.
> 2. $H_a$ claims the presence of something hence considered Positive, for example: presence of cancer, being guilty, etc.

#### Type-1 error vs Type-2 error

Between Type-1 and Type-2 errors which one should be low?

This depends on the domain. For example in:

1. **Judicial** process should have **very low Type-1 error** because an innocent should not be punished for the crime that they didn't commit.
2. **Medical** diagnostic should have **very low Type-2 error** because a disease should not go undetected, additional tests can be performed to reaffirm. 

#### Summary

<div style="display: inline-block">

| Type-1 Error                 | Type-2 Error                     |
| :--------------------------- | :------------------------------- |
| Alpha $\alpha$               | Beta $\beta$                     |
| False Positive               | False Negative                   |
| Incorrect rejection of $H_0$ | Failed to reject incorrect $H_0$ |

</div>

### 2.2 Power of test

In a Machine learning system the goal is to **reduce both Type-1 and Type-2 errors**.

#### Definition

1. Power represents the probability of correctly rejecting a False Null Hypothesis.
2. $1 - \beta$ (True Positive) is also called as **Power of test**.

```
                        Actual
                    0           1         
             +------------+------------+
           0 | 1 - α (TN) |     β (FN) |
Predicted    +------------+------------+
           1 |     α (FP) | 1 - β (TP) |
             +------------+------------+
```

The goal of power of test is to reduce Type-2 error $(\beta)$.


$\Rightarrow$ $\text{FN} + \quad\;\;\text{TP} = 1$  
$\Rightarrow$ $\;\;\;\beta + (1 - \beta) = 1$


1. To reduce Type-2 error $\beta$ (False Negative) we have to increase $1 - \beta$ (True Positive).
2. To reduce Type-1 error $\alpha$ (False Positive) we have to increase $1 - \alpha$ (True Negative).

> **Note**:
>
> To increase Power of test $(1 - \beta)$ we have to reduce $\beta$.

#### Relationship between Alpha and Beta

$
\begin{align}
\large
\alpha = \frac{1}{\beta}
\end{align}
$

Since $\alpha$ and $\beta$ are inversely proportional to each other, mathematically it is not possible to reduce both the errors at the same time.  
We always need to find a trade off  between $\alpha$ and $\beta$.

#### Factors influencing Power

1. Sample size $(n)$: As sample size increases power increases.
2. Significance level $(\alpha)$: As Significance level increases beta reduces (hence power increases).
3. Variance $(\sigma)$: As Variance increases Type-2 error increases (hence power reduces).
4. Effect Size $(d)$: As effect size increases power increases.

#### Summary

<div style="display: inline-block">

|   # | Metric         |       Symbol | Representation     | Error Type   | Interpretation in Hypothesis Testing                        |
| --: | :------------- | -----------: | :----------------- | :----------- | :---------------------------------------------------------- |
|   1 | True Negative  | $1 - \alpha$ | Confidence level   |              |                                                             |
|   2 | False Negative |      $\beta$ |                    | Type-2 Error | Failing to reject an incorrect (False) Null Hypothesis.     |
|   3 | True Positive  |  $1 - \beta$ | Power of Test      |              | Probability of correctly rejecting a False Null Hypothesis. |
|   4 | False Positive |     $\alpha$ | Significance level | Type-1 Error | Incorrect rejection of a True Null Hypothesis.              |

</div>

## 3 Types of Tailed Test

Types of tests:

1. One Tailed Test
    1. Left Tailed Test
    2. Right Tailed test
2. Two Tailed test

### 3.1 Cases

#### Case #1

$H_0: \mu \ge value$  
$H_a: \mu \lt value$  

Perform Left-tailed test.

#### Case #2

$H_0 : \mu \le value$  
$H_a : \mu \gt value$  

Perform Right-tailed test.

#### Case #3

$H_0 : \mu = value$  
$H_a : \mu \ne value$  

Perform Two-tailed test.

> **Note**:
>
> Equal comparison i.e., $=$ operator will always be present in $H_0$

1. if condition is $<$ or $\le$ in $H_a$ then perform **Left-Tailed** test.
2. if condition is $>$ or $\ge$ in $H_a$ then perform **Right-Tailed** test.
3. if condition is $\ne$ in $H_a$ then perform **Two-Tailed** test.

#### Summary

<div style="display: inline-block">

| Case         | Tail test type    |
| :----------: | :---------------- |
| $<$ or $\le$ | Left tailed test  |
| $>$ or $\ge$ | Right tailed test |
| $\ne$        | Two tailed test   |

</div>

## 4 Hypothesis Testing Framework

### 4.1 Steps in Hypothesis Testing

- **STEP #1**: Formulate Null-Hypothesis $H_0$ and Alternate-Hypothesis $H_a$
- **STEP #2**: Choose a Distribution, Test-statistic and Significance-level $(\alpha)$
- **STEP #3**: Select type of test: Left / Right / Two-tailed
- **STEP #4**: Compute $\text{p-value}$
- **STEP #5**: Compare $\text{p-value}$ with $\alpha$, reject $H_0$ if $\text{p-value} \le \alpha$ else reject $H_a$

### 4.2 Examples

#### Coin Toss

##### STEP 1: Formulate Null and Alternate Hypothesis

- $H_0$ : Coin is fair
- $H_a$ : Coin is biased

Above formulation can be mathematically written as,

- $H_0 : p = 0.5$
- $H_a : p \ne 0.5$

##### STEP 2: Choose Significance level

````python
# Distribution: Binomial distribution
# Significance level: 0.05
alpha = 0.05
````

##### STEP 3: Select type of test

Since $H_a$ has $\ne$ symbol perform two tailed test.

##### STEP 4: Compute p-value

````python
p_value = 0.001  # Dummy value for example
````

##### STEP 5: Compare p-value with alpha

````python
if p_value < alpha:
    print("Reject Null-Hypothesis i.e., accept Alternate-Hypothesis.")
else:
    print("Failed to reject Null-Hypothesis i.e., reject Alternate-Hypothesis.")
````
