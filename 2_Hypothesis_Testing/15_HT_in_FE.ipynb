### Skewness

1. Skewness measures symmetry.
2. Skewness describes **shape of data**.
3. Normal distribution has a Skewness value of 0.
4. Three types of Skewness:
    1. Positive
    2. Neutral
    3. Negative

### Kurtosis

1. Kurtosis measures the sharpness of the peak (thickness of tails).
2. Kurtosis helps in detecting outliers in dataset.
3. Kurtosis describes the **concentration of data-points**.
4. Normal distribution has a kurtosis value of 3

### Excess Kurtosis

1. Excess Kurtosis := kurt_data - kurt_normal
2. Normal distribution has an excess kurtosis value of 0
3. Three types of Excess Kurtosis:
    1. Leptokurtic
    2. Mesokurtic
    3. Platykurtic

#### Leptokurtic

1. Kurtosis > 3
2. Excess kurtosis is positive
3. Longer tails and higher chances of outliers on both ends

#### Mesokurtic

1. Kurtosis ~ 3
2. Excess kurtosis ~ zero
3. Nearly normal

#### Platykurtic

1. Kurtosis < 3
2. Excess kurtosis is negative
3. Most data points around mean

### Outlier detection code

```python
import numpy as np
```

```python
# Given dataset
data = np.array([15, 18, 22, 23, 25, 28, 40, 75, 200])

# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Determine the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = []
# Identify outliers
for value in data:
    if value < lower_bound or value > upper_bound:
        outliers.append(value.item())

# Print the results
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Identified Outliers:", outliers)
```
