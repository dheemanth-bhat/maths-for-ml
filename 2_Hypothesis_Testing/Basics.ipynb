## 1 Hypothesis Testing

### 1.1 Basic terminologies

Process of designing some experiments, collecting data and evidence to prove a case (measure the impact) in point.

There can be three outcomes of impact:

1. No Impact
2. Positive Impact
3. Negative Impact

1. Assumptions
2. Data / Evidence to support or negate the assumption.

1. Default Assumptions
2. Claim

> **Note**:
> 
> Burden of proof is on those who want to reject default assumption.

When there is enough data or evidence to backup the claim then the default assumption is rejected.

### 1.2 Null vs Alternate Hypothesis

#### Definition

1. Default assumption is called as $H_0$ - Null Hypothesis
2. Opposite of Default assumption is called as $H_a$ - Alternate Hypothesis

Other terms used for Default Assumptions:

1. Null Hypothesis $H_0$
2. Status Quo
3. Baseline Assumptions

Hypothesis Testing will have two outcomes:

1. Successfully reject $H_0$ or accept $H_a$
2. Successfully reject $H_a$ or failed to reject $H_0$

> **Note**:
>
> 1. There is no term or process to **accept Null Hypothesis**.
> 2. We gather evidence towards our claim to either **reject Null Hypothesis** or **failed to reject Null Hypothesis**.
> 3. We gather evidence towards our claim to either **accept Alternate Hypothesis** or **reject Alternate Hypothesis**.

### 1.3 p-value

#### Definition

$\text{p-value}$ is probability of observing data as extreme as what was observed given that Null Hypothesis $H_0$ is true.

We reject $H_0$ when $\text{p-value}$ is very low.

> **Note**:
> 
> p-value is not probability value. p-value is neither the probability of $H_0$ nor $H_a$.

#### Formula

$\text{p-value} = P(data \mid H_0)$

### 1.4 Significance level and Confidence level

#### Definition

$
\begin{array}{l}
\text{if p-value} < \alpha \text{:} \\[2pt]
\quad \text{Reject $h_0$ i.e., Accept $h_a$} \\[8pt]
\text{else:} \\[2pt]
\quad \text{Failed to reject Reject $h_0$ i.e., Reject $h_a$}
\end{array}
$

1. $\alpha$ is called as **Significance level**
2. $1 - \alpha$ is called as **Confidence level**

> **Note**:
>
> Generally in most of the industries 95% confidence level is considered hence  
> most common value for Confidence Level is 0.95 and Significance Level is 0.05

1. Value of $\alpha$ depends on the domain of the problem.
2. For example medical domain can have very low threshold of 0.1
3. Generally the value of $\alpha$ must be less than 0.5

#### Risks involved in Significance level

```
                      Actual
                   0         1     
             +----------+----------+
           0 | No Error |  Error   |
Predicted    +----------+----------+
           1 |  Error   | No Error |
             +----------+----------+
```

> **Note**:
>
> Above matrix is called as confusion matrix.

### 1.5 Types of Error

1. Type 1 Error $(\alpha)$ - False Positive - Significance level
2. Type 2 Error $(\beta)$ - False Negative

#### Example

### 1.6 Types of Tailed Test

1. One Tailed Test
    1. Left Tailed Test
    2. Right Tailed test
2. Two Tailed test

1. $<$ or $\le$ - Left tailed test
2. $>$ or $\ge$ - Right tailed test
3. $\ne$ - Two tailed test

**STEP 1**: Assumption  
**STEP 2**: Data/Evidence  
**STEP 3**: Calculate p-value  
**STEP 4**: Threshold  

### 1.7 Hypothesis Testing Framework

1. Setup the $H_0$ and $H_a$
2. Choose the distribution
3. Select the Tail type: Left / Right / Two-Tailed test
4. Compute p-value
5. p-value < $\alpha$ then Reject $H_0$

### 1.8 Examples

#### Quiz #1

The juice brand claims that the mean sugar content in its juice boxes, produced using a new manufacturing process, is less than 8 grams.  
The Food Safety and Standards Authority of India (FSSAI) wants to test the null hypothesis in this scenario.   
Set up the null (H0) and alternate (H1) hypotheses for this claim, and choose the correct option:

1. H0: The sugar content in the juice boxes produced using the new manufacturing process is not equal to 8 grams.  
   H1: The sugar content in the juice boxes produced using the new manufacturing process is equal to 8 grams.
2. H0: The sugar content in the juice boxes produced using the new manufacturing process is greater than 8 grams.  
   H1: The sugar content in the juice boxes produced using the new manufacturing process is less than 8 grams.
3. H0: The sugar content in the juice boxes produced using the new manufacturing process is less than 8 grams.  
   H1: The sugar content in the juice boxes produced using the new manufacturing process is greater than 8 grams.
4. H0: The sugar content in the juice boxes produced using the new manufacturing process is greater than or equal to 8 grams.  
   H1: The sugar content in the juice boxes produced using the new manufacturing process is less than 8 grams.

##### Answer

4. H0: The sugar content in the juice boxes produced using the new manufacturing process is greater than or equal to 8 grams.  
   H1: The sugar content in the juice boxes produced using the new manufacturing process is less than 8 grams.

#### Quiz #2

Local people claim that the river water quality is good.  
An environmental agency wants to test their claim. Upon testing, they obtained a p-value of 0.001.  
The significance level ($\alpha$) is set at 0.05.  

What is the appropriate conclusion?

1. Reject the null hypothesis
2. Fail to reject the null hypothesis
3. Increase the significance level
4. Conduct additional tests

##### Answer

1. Reject the null hypothesis

#### Quiz #3

A restaurant says that its new menu item has a mean waiting time of less than equal to 10 minutes.  
You claim that the mean wait time is greater than 10 minutes.  
What should be the null and alternate hypotheses to test their claim?

1. $H_0: \mu \le 10 \text{ and } H_a: \mu > 10$
1. $H_0: \mu = 10 \text{ and } H_a: \mu > 10$
1. $H_0: \mu = 10 \text{ and } H_a: \mu \ne 10$
1. $H_0: \mu \gt 10 \text{ and } H_a: \mu \le 10$

##### Answer

1. $H_0: \mu \le 10 \text{ and } H_a: \mu > 10$
