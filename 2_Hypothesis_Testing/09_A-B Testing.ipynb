```python
import numpy as np
import pandas as pd
```

```python
yt_df = pd.read_csv("../0_data/06_youtube/ab_test_data.csv")

row_count, col_count = yt_df.shape

display(yt_df.head(3))
print(f"Dataset has {row_count} rows and {col_count} columns.")
```

```python
yt_ct = (
    pd.crosstab(
        index=yt_df["premium"],
        columns=yt_df["customer_segmnt"],
        margins=True,
        margins_name="Total",
        normalize=True,
    )
    .multiply(100)
    .round(2)
)
yt_ct
```
