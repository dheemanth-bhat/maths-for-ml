## 1 Dimensionality Reduction

### 1.1 Dimensionality Curse

#### What is Curse of Dimensionality?

- Too many dimensions or features or columns in the dataset.

#### What is Dimensionality reduction?

Given a dataset $D$ where $D$ is defined as:

$$
\large
\begin{aligned}
D := \biggl\{\bigl(x_i, y_i \bigr)\;;\;x\in\mathbb{R}^d\;;\;y\in\{-1,+1\}\biggr\}_{i=1}^n
\end{aligned}
$$

Where,

- $D$ is Dataset
- $x$ are features of datatype Real numbers $\mathbb{R}$
- $y$ are labels
- $n$ is the total number of rows in the dataset
- $d$ is the total number of features/columns (dimension) in the dataset

The goal of Dimensionality reduction is to reduce dimensions of data to $d^\prime$ where $d^\prime < d$.

### 1.2 Dimensionality Reduction Techniques

Different techniques used in Dimensionality Reduction

#### 1 Feature Engineering

- Combining two or more features into single feature using some basic **formula**.
- It is generally done without loosing any valuable information present in the dataset.
- For example combining height and weight into BMI. Formula for $\displaystyle \text{BMI} = \frac{\text{weight}}{\text{height}^2}$

#### 2 Deleting Features

- Deleting useless features/columns in the dataset.
- It is **not recommended** to use this technique as it can result in **information loss**.

#### 3 Principal Component Analysis (PCA)

- Complex mathematical technique to reduce dimensions in dataset when features cannot be merged using some basic formula.

### 1.3 Correlation

![Correlation](images/correlation.png)

- Correlation is a measure of relationship between two features.
- There are two types of Correlation:
   1. Positive Correlation
   2. Negative Correlation
- PCA uses Correlation in dimensionality reduction.

#### Covariance Matrix

##### Formula

$$
\large
\begin{aligned}
\text{Correlation} = \frac{\text{Covariance}(x, y)}{\vert{x}\vert\;\vert{y}\vert}
\end{aligned}
$$

1. Covariance can range between $[-\infty, +\infty]$
2. Correlation is Covariance scaled between $[-1, +1]$

## 2 Principal Component Analysis (PCA)

### 2.1 How PCA Works?

PCA performs two steps to reduce dimensionality of data:

1. Rotating and/or Shifting axis (in the direction of maximum variance).
2. Projection of points

#### 1 Rotating and/or Shifting axis

![PCA Techniques](images/PCA_techniques.png)

- Two features are plotted on a Cartesian plane visualizing its correlation.
- The Cartesian plane is rotated/shifted in such a way that the **maximum variance is preserved on one of the axis**.
- Axis with maximum variance is called as **Principal Component 1** (PC1).
- The other axis is automatically perpendicular to Principal Component 1 and is called as **Principal Component 2** (PC2).

#### 2 Projection of points

![PCA Projection of Points](images/PCA_projection_of_points.png)

- Variance on Principal Component 2 will be very minimal.
- Coordinates spread across Principal Component 1 will be **projected** on to that axis.
- Principal Component 2 can now be completely ignored focusing only on Principal Component 1.
- Hence the dimensionality of the data is reduced from 2 to 1 by retaining maximum variance on one of the axis..

> **Note**:
>
> There will be some information loss when projecting points on the Principal Component 1.

### 2.2 Issues with PCA

PCA does not work when:

- There is no correlation between two features.
- Strongly non-linear relationships.

### 2.3 PCA Linear Equation

#### Formula

Linear equation for PCA:

$$
\large
\begin{aligned}
\text{feature}_{\text{new}} = u_1*\text{feature}_1\;+\;u_2*\text{feature}_2\;+\;\ldots\;+\; u_n*\text{feature}_n
\end{aligned}
$$

Where $\text{feature}_{\text{new}}$ is a linear combination of features: $\{\text{feature}_1, \text{feature}_2,\;\ldots,\;\text{feature}_n\}$

#### Properties

- Goal of PCA is to find that **optimal vector**: $[u_1 u_2 \dots u_n]$ that **maximizes the variance**.

### 2.4 Shifting Axis

#### How Shifting of Axis works?

- Shifting of axis is indirectly achieved using feature standardization process.
- Features are mean centered i.e., each value is subtracted with the mean value of the same feature.

### 2.5 Optimization of Variance in Projection

#### Variance

Formula of variance:

$$
\large
\begin{align}
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
\end{align}
$$

#### Projection of Vectors

Formula to calculate projection of $\vec{x}$ on $\vec{u}$:

$$
\large
\begin{aligned}
\text{Projection of $\vec{x}$ on $\vec{y}$} = \frac{\vec{x}\cdot\vec{u}}{\Vert{u}\Vert}
\end{aligned}
$$

Combining above to equations to formulate an equation for optimization/maximizing variance during projection of vectors:

$$
\large
\begin{aligned}
\vec{u}^* = \arg\max_{\overrightarrow{u}}\;\frac{1}{n}\sum_{i=1}^n\Biggl(\frac{\vec{x_i}^T \cdot u_i}{\Vert{u}\Vert} \;-\;\mu \Biggr)^2
\end{aligned}
$$

#### Constraint Optimization

##### Constraint

Equation for maximizing variance during projection of vectors with constraint enforced on it:

$$
\large
\begin{aligned}
\vec{u}^* = \arg\max_{\overrightarrow{u}}\;\frac{1}{n}\sum_{i=1}^n\Biggl(\frac{\vec{x_i}^T \cdot \vec{u_i}}{\Vert{\vec{u}}\Vert} \;-\;\mu \Biggr)^2 \quad \text{such that}:\;\Vert{\vec{u}}\Vert = 1
\end{aligned}
$$

Equation for maximizing variance during projection of vectors with constraint optimization:

$$
\large
\begin{aligned}
\vec{u}^* = \arg\max_{\overrightarrow{u}}\;\frac{1}{n}\sum_{i=1}^n\Bigl(\vec{x_i}^T \cdot \vec{u_i}\Bigr)^2\;+\;\lambda\bigl(\Vert{\vec{u}}\Vert - 1\bigr)
\end{aligned}
$$

### 2.6 Principal Components

#### Principal Component Matrix

$$
\large
\begin{aligned}
\underset{\displaystyle n \times d}{
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1d} \\[3pt]
\vdots & \vdots & \vdots & \vdots \\[3pt]
x_{n1} & x_{n2} & \cdots & x_{nd} \\[3pt]
\end{bmatrix} \\[25pt]
}
\;\;\cdot\;\;
\underset{\displaystyle d \times d}{
\begin{bmatrix}
u_{11} & u_{12} & \cdots & u_{1d} \\[3pt]
\vdots & \vdots & \vdots & \vdots \\[3pt]
u_{d1} & u_{d2} & \cdots & u_{dd} \\[3pt]
\end{bmatrix} \\[25pt]
}
\;\;=\;\;
\underset{\displaystyle n \times d}{
\begin{bmatrix}
v_{11} & v_{12} & \cdots & v_{1d} \\[3pt]
\vdots & \vdots & \vdots & \vdots \\[3pt]
v_{n1} & v_{n2} & \cdots & v_{nd} \\[3pt]
\end{bmatrix} \\[25pt]
}
\end{aligned}
$$

$$
\large
\begin{aligned}
X_{n \times d}^T \cdot u_{d \times d} = v_{n \times d}
\end{aligned}
$$

#### Properties

- If we have $k$ features we will get $k$ Principal Components (new features).
- Principal Components are **Orthogonal**.
- Principal Components are ordered on importance (Eigen values).

## 3 Eigen Decomposition

### 3.1 What is Eigen Decomposition?

If a Covariance matrix $S$ exists in such a way that it satisfies the condition: $S\cdot\vec{u}=\lambda\cdot\vec{u}$  
then, $\vec{u}$ is called the Eigen vector for $S$ and $\lambda$ is called as its Eigen value.

#### Properties

- Eigen values will be a **Diagonal Matrix**.
- Optimal values for $u$ vector can be computed using Eigen Decomposition.
- While optimal values for $u$ vector can be arrived at using Gradient Ascent/Descent algorithm, alternate approach is Eigen Decomposition

#### Eigen Values

$$
\large
\begin{aligned}
\underset{\displaystyle n \times d}{
\begin{bmatrix}
e_{11} & 0      & \cdots & 0      \\[3pt]
0      & e_{21} & \cdots & 0      \\[3pt]
\vdots & \vdots & \ddots & \vdots \\[3pt]
0      & 0      & \cdots & e_{mn} \\[3pt]
\end{bmatrix}
}
\end{aligned}
$$

##### Properties

- Eigen Values are used to calculate **proportion of variance**.

#### Eigen Vector

##### Properties

- Eigen Vectors are sorted in the order of Eigen Values.
