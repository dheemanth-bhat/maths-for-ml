## 1 Distances

### 1.1 Properties of Lines

#### Shifting Lines

The probability that a data point belongs to the correct class is directly proportional to the distance of a point from the decision boundary.

1. Points closer to the Decision Boundary have lower probability of being in their class.
2. Points closer to the Decision Boundary have higher probability of being in their class.
3. Points lying on the Decision Boundary have 0.5 probability of belonging to either class.

#### Orthogonality

Weight vector is always perpendicular to Decision Boundary.

### 1.2 Distance of Line from Origin

#### Formula

Formula to calculate distance of a vector $\vec{w}$ from origin:

$$
\begin{aligned}
\large
d = \frac{-w_0}{\Vert{w}\Vert}
\end{aligned}
$$

#### Properties

The minus sign in the formula shows the direction. When the distance is:

1. Negative: Decision boundary and the vector are in the **opposite direction** from origin.
2. Positive: Decision boundary and the vector are in the **same direction** from origin.

### 1.3 Distance of Point from Line

#### Formula

Formula to calculate the distance between a point $x_0$ and line (Decision Boundary) using vector $\vec{w}$:

$$
\begin{aligned}
\large
d = \frac{w^{T}x_0 + w_0}{\Vert{w}\Vert}
\end{aligned}
$$

#### Examples

##### Code

```python
import numpy as np
```

```python
def magnitude(v):
    """
    Function to calculate Euclidean distance
    (magnitude) of a vector `v` from origin.
    """
    return np.sqrt(np.sum(np.asarray(v) ** 2))
```

```python
def distance_bw_point_line(w, x, w_0):
    """
    Function to calculate distance between
    data-point `x` and the Decision Boundary.
    """
    dot_p = np.dot(np.asarray(w), np.asarray(x))
    print("Dot product:", dot_p)

    mag = magnitude(w)
    print("Magnitude:", round(mag, 6))

    dist = (dot_p + w_0) / mag
    return round(dist, 6)
```

##### Example #1

Find the distance of the point (3, 9) from the plane $-2x + 3y -3 = 0$?

```python
d1 = distance_bw_point_line(w=[-2, 3], x=[3, 9], w_0=-3)
print("Distance:", d1)
```

##### Example #2

Find the distance of the point (6, 2) from the plane $-2x + 3y -3 = 0$?

```python
d2 = distance_bw_point_line(w=[-2, 3], x=[6, 2], w_0=-3)
print("Distance:", d2)
```

##### Example #3

Find the distance of the point (6, 5) from the plane $-2x + 3y -3 = 0$?

```python
d3 = distance_bw_point_line(w=[-2, 3], x=[6, 5], w_0=-3)
print("Distance:", d3)
```

##### Example #4

Find the distance of the point (-6, 0, 0) from the plane $2x - 3y + 6z = 2$?

```python
d4 = distance_bw_point_line(w=[2, -3, 6], x=[-6, 0, 0], w_0=-2)
print("Distance:", d4)
```

### 1.4 Distance between two parallel lines

#### Formula

To calculate distance between two parallel planes, divide the difference of their constants by their normal vectors' magnitude.

$$
\begin{aligned}
\large
d = \frac{{w_0}^{(1)} - {w_0}^{(2)}}{\Vert{w}\Vert}
\end{aligned}
$$

#### Code

```python
def distance_bw_lines(w_01, w_02, w):
    """
    Function to calculate distance between two lines.
    """
    mag = magnitude(w)
    dist = (w_01 - w_02) / mag

    return round(dist, 6)
```


#### Examples

##### Example

Calculate distance between $w^T{x} + 5 = 0$, $w^T{x} + 3 = 0$ where $w = [3, 2, 1]$.

```python
d1 = distance_bw_lines(w_01=5, w_02=3, w=[3, 2, 1])
print("Distance:", d1)
```
