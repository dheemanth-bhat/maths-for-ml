## 2 Slope-Intercept

### 2.1 Slope-Intercept Equation

$$
\large
\begin{aligned}
y = m \cdot x \;+\; c
\end{aligned}
$$

Where,

- $m$ - Slope (angle between straight line and x-axis)
- $c$ - intercept (distance of straight line from the origin)
- $x$ - Independent variable
- $y$ - Dependent variable

### 2.2 Slope

Given two points $(x_1, y_1)$ and $(x_2, y_2)$, formula to calculate Slope $m$:

$$
\large
\begin{aligned}
m = \frac{y_2 - y_1}{x_2 - x_1}
\end{aligned}
$$

### 2.3 Intercept

Given two points $(x_1, y_1)$ and $(x_2, y_2)$, formula to calculate Intercept $c$:

$$
\large
\begin{aligned}
c = y_1 - (m \cdot x_1) \\[5pt]
\text{or} \\[5pt]
c = y_2 - (m \cdot x_2)
\end{aligned}
$$

### 2.4 Alternate Line Equation

$$
\large
\begin{equation}
w_1 x + w_2 y + w_0 = 0
\end{equation}
$$

Rearranging above equation (1) in Slope-Intercept form: 

$$
\large
\begin{aligned}
y = \biggl(-\;\frac{w_1}{w_2}\biggr)x_1 + \biggl(-\;\frac{w_0}{w_2}\biggr)
\end{aligned}
$$

Where Slope and Intercept can be derived as, 

$$
\large
\begin{aligned}
m = -\;\frac{w_1}{w_2}
\end{aligned}
$$

$$
\large
\begin{aligned}
c = -\;\frac{w_0}{w_2}
\end{aligned}
$$

### 2.5 Geometric Relationship b/w lines

#### Parallel Lines

When given two lines $y = m_1 x + c_1$ and $y = m_2 x + c_2$ are parallel:

$$
\large
\begin{aligned}
m_1 = m_2
\end{aligned}
$$

#### Perpendicular Lines (Orthogonality)

When given two lines $y = m_1 x + c_1$ and $y = m_2 x + c_2$ are perpendicular:

$$
\large
\begin{aligned}
m_1 \times m_2 = -1
\end{aligned}
$$

## 3 General Line Equation

### 3.1 Line Equation (2D Hyperplane)

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 = 0
\end{aligned}
$$

### 3.2 Plane Equation (3D Hyperplane)

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_3 x_3 + w_0 = 0
\end{aligned}
$$

### 3.3 nD Hyperplane Equation

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + \ldots +  w_n x_n + w_0 = 0
\end{aligned}
$$

## 4 Binary Classification

### 4.1 Mapping Function

$$
\large
\begin{aligned}
y = f(x) = w_1 x_1 + w_2 x_2 + w_0
\end{aligned}
$$

### 4.2 Positive Half-space

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 > 0
\end{aligned}
$$

### 4.3 Negative Half-space

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 < 0
\end{aligned}
$$

### 4.4 Decision Boundary

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 = 0
\end{aligned}
$$
