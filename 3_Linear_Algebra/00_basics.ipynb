## 1 Trigonometry

### 1.1 Trigonometric Ratios

![Trigonometric Ratios](images/trigonometric_ratios.png)

$$
\large
\begin{aligned}
\sin \uptheta = \frac{\text{Opposite}}{\text{Hypotenuse}}
\end{aligned}
$$

$$
\large
\begin{aligned}
\cos \uptheta = \frac{\text{Adjacent}}{\text{Hypotenuse}}
\end{aligned}
$$

$$
\large
\begin{aligned}
\tan \uptheta = \frac{\text{Opposite}}{\text{Adjacent}}
\end{aligned}
$$
