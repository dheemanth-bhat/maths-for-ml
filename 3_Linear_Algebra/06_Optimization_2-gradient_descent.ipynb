## 1 Differentiation

### 1.1 Geometric Representation

#### Slope of a Line

![Slope of a line](images/slope_of_line1.png)

Formula to calculate slope of a line intersecting two points $(x_i, y_1)$ and $(x_2, y_2)$

$$
\large
\begin{aligned}
m &= \frac{y_2 - y_1}{x_2 - x_1} \\[12pt]
  &= \frac{\Delta\;y}{\Delta\;x}
\end{aligned}
$$

- $\Delta\;x$ is change in x
- $\Delta\;y$ is change in y
- As $x$ increases if $y$ is increasing then slope is in positive.
- As $x$ increases if $y$ is decreasing then slope is in negative.

#### Slope of Secant Line

![Slope of curve](images/slope_of_curve1.png)

Formula to calculate slope of a curve intersecting two points $(x_i, y_1)$ and $(x_2, y_2)$

$$
\large
\begin{aligned}
m &= \frac{y_2 - y_1}{x_2 - x_1} \\[12pt]
  &= \frac{f(x_2) - f(x_1)}{x_2 - x_1}
\end{aligned}
$$

### 1.2 Derivative

#### Secant vs Tangent

- If a straight line intersects a curve at two points its called as **Secant**.
- If a straight line intersects a curve at only one point its called as **Tangent**.

#### Slope of Tangent Line

![Slope of curve derivative](images/slope_of_curve_derivative1.png)

Formula to calculate slope of a curve intersecting at only one point $(x_i, y_1)$

$$
\large
\begin{aligned}
f^{\prime}(x) = m = \lim_{\Delta{x} \to 0} \frac{f(x_1 + \Delta{x}) - f(x_1)}{\Delta{x_1}}
\end{aligned}
$$

#### Derivative of a Function

**Q**: What does the derivative of a function represent?

**Ans**:

The slope of the tangent line to the function's graph at a given point.

$$
\large
\begin{aligned}
m &= \frac{f(x_1 + \Delta\;x) - f(x_1)}{(x_1 + \Delta\;x) - x_1} \\[12pt]
  &= \frac{f(x_1 + \Delta\;x) - f(x_1)}{\cancel{x_1} + \Delta\;x - \cancel{x_1}} \\[12pt]
  &= \frac{f(x + \Delta\;x) - f(x)}{\Delta\;x}
\end{aligned}
$$

#### Critical Point

Slope will be zero at Critical Point.

#### Double Derivative

$$
\large
\begin{aligned}
f^{\prime\prime}(x)
\begin{cases} 
f^{\prime\prime}(x) > 0 \quad \text{Minima} \\[8pt]
f^{\prime\prime}(x) < 0 \quad \text{Maxima} \\[8pt]
f^{\prime\prime}(x) = 0 \quad \text{Saddle point} 
\end{cases}
\end{aligned}
$$

#### Summary

1. Using derivatives, identify Critical points (where slope is zero).
2. Critical points can be maxima or minima.
3. Using double derivatives, identify maxima/minima.
   - If double derivative is **greater than zero** then the critical point is **Minima**.
   - If double derivative is **less than zero** then the critical point is **Maxima**.
   - If double derivative is **equal to zero** then the critical point is **Saddle point**.
