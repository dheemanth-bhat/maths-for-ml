### 1.1 Mathematics for ML

Five important branches of mathematics used in ML:

1. Statistics
2. Probability Theory
3. Linear Algebra
4. Coordinate Geometry
5. Calculus

### 1.2 Key terms

#### Data / Dataset

1. Dataset contains **data-points** i.e., each row/observation in the dataset is a data-point.
2. ML Datasets have two components $X$ (multiple columns) and $Y$ (single column).
3. Datasets with two components $X$ and $Y$ are called as historical (or labeled) dataset.
4. Each data-point can be represented as a coordinate on a Cartesian coordinate system (for example Cartesian plane).

#### X vs Y

```
X - Features | Input Variables | Independent variables | Predictors
Y - Label    | Target Variable | Dependent variable    | Predicted variables
```

#### Features

1. Features are **Independent Variables**.
2. Each feature is a **dimension** on a Cartesian Plane.
3. Two dimensional data can be visualized using **scatter plot**.
4. Techniques like **TSNE**, **PCA** are used for visualizing higher dimensional data.

#### Label

1. Label is a **Dependent Variable**.
2. A Label identifies to which class a data-point belong to.
3. A data-point can have at most one label.
4. There can be two or more types label in a dataset.

#### Classifier

- A Classifier segregates the data-points into two or more groups (depending on the number of classes).  
- A classifier is also called as Separator. 

There are different types of Separators:

1. Liner Separator: A straight line separating data-points belonging to two classes.
2. Non-liner Separator: A complex separator for non-linear data.

##### Types of Classification

Based on the number of types of labels in the dataset, there are two types of classification problems:

1. Binary Classification: Two types of labels in the dataset.
2. Multi-class classification: Two or more types of labels in the dataset.

### 1.4 Training Pipeline

A Training pipeline is a cyclic process that contains below three components resulting in a ML Model.

1. Mapping Function
2. Loss Function
3. Optimization

#### 1 Mapping Function

##### Definition

Mapping function := A function $f$ that takes $x$ as an input gives $y$ as an output.

Mathematically a mapping function can be represented as:

$$
\large
\begin{aligned}
y = f(x)
\end{aligned}
$$

##### Properties

1. Initially $f(x)$ predicts values randomly hence we need a loss function.
2. Concepts from Linear Algebra and Coordinate Geometry are used in constructing a mapping function.

#### 2 Loss Function

##### Definition

Loss function := Function that measures how close the predicted value from $f(x)$ is to the actual label $y$.  

##### Properties

1. Loss function only measures how "wrong" (delta) the predicted value is from the actual value, we need optimization to improvise the loss function.
2. Value from a Loss function is just a metric.

#### 3 Optimization

##### Definition

Optimization := Process to improvise the loss function.

##### Properties

1. Concepts like maximization, minimization from Calculus are used in optimizing a Loss function.
2. Optimizing a classifier is like finding a **best fit line** that segregates the data-points based on their class labels.
3. Here a classifier is an ML Model or a line or a mathematical equation.
4. Optimization helps to generalize the model for future unseen data (rather than exactly fitting a line on a dataset).

#### What is an ML Model?

1. An ML model is a parameterized mathematical function that maps input features to predicted targets.

### 2.2 Slope-Intercept Equation (SIE)

Slope-Intercept form of a Linear Equation a.k.a equation of straight line.

#### Decision Boundary

1. Decision Boundary := Line that segregates data-points (say on a scatter plot) into different classes/labels.
2. Binary classification has **one Decision Boundary**.
3. Multi-class classification has **many Decision Boundaries**.

##### Equation #1

For given point $(x, y)$, equation of a single straight line a.k.a Slope intercept equation:

$$
\large
\begin{aligned}
y = m \cdot x \;+\; c
\end{aligned}
$$

Where,

- $m$ - Slope (angle between straight line and x-axis)
- $c$ - intercept (distance of straight line from the origin)
- $x$ - Independent variable
- $y$ - Dependent variable

#### Slope

##### Properties

1. In Slope-Intercept equation $m$ represents the Slope.
2. Slope determines the angle $\uptheta$ between the Decision Boundary and the x-axis.

![Slope Intercept](images/slope_intercept.png)

##### Formula

If we have two points $(x_1, y_1)$ and $(x_2, y_2)$ that forms a line, then Trigonometric method to calculate slope between two lines:

$$
\large
\begin{aligned}
m = \tan \uptheta = \frac{y_2 - y_1}{x_2 - x_1}
\end{aligned}
$$

or,

$$
\large
\begin{aligned}
m = \tan \uptheta = \frac{\Delta\;y}{\Delta\;x}
\end{aligned}
$$

##### Correlation between x and y

- if both $\Delta x$ and $\Delta y$ are changing in the same direction then its positive correlation, hence **slope is positive**.
- if $\Delta x$ and $\Delta y$ are changing in the opposite direction then its negative correlation, hence **slope is negative**.
- if $\Delta x = 0$ then the line is vertical and slope is $\infty$ or **slope is undefined**.
- if $\Delta y = 0$ then the line is horizontal and **slope is zero**.

- Positive Slope: https://www.desmos.com/calculator/blxhp7ey9x
- Negative Slope: https://www.desmos.com/calculator/shdewjcjxv

#### Intercept

##### Properties

1. In Slope-Intercept equation $c$ represents the Intercept.
2. Intercept shifts the Decision Boundary away from the origin $(0, 0)$ on a cartesian plane.
3. When Intercept is zero Decision Boundary passes through the origin.

##### Formula

If we have two points $(x_1, y_1)$ and $(x_2, y_2)$ that forms a line, then formula to calculate intercept:

$$
\large
\begin{aligned}
c = y_1 - (m \cdot x_1) \\[5pt]
\text{or} \\[5pt]
c = y_2 - (m \cdot x_2)
\end{aligned}
$$

#### Issues with Slope-Intercept Equation

1. While using slope-intercept equation, when $\Delta x = 0$, decision boundary will be $90^{\circ}$ to the x-axis.
2. A decision boundary with $90^0$ to x-axis has slope $m = \text{undefined}$ because $\tan (90^{\circ})$ is $\infty$ or undefined.
3. But a vertical line is a valid decision boundary segregating data-point into two half-spaces to its left and right.
5. Hence an alternate equation for straight line called: **General Line Equation** is used instead of Slope-Intercept equation.

#### Alternate Equation for Straight line

##### Equation #2

$$
\large
\begin{aligned}
w_1 x + w_2 y + w_0 = 0
\end{aligned}
$$

##### Equation #3

Rearranging Equation #2 as Equation #3: 

$$
\large
\begin{aligned}
y = \biggl(-\;\frac{w_1}{w_2}\biggr)x_1 + \biggl(-\;\frac{w_0}{w_2}\biggr)
\end{aligned}
$$

Where Slope and Intercept can be derived as, 

$$
\large
\begin{aligned}
m = -\;\frac{w_1}{w_2}
\end{aligned}
$$

$$
\large
\begin{aligned}
c = -\;\frac{w_0}{w_2}
\end{aligned}
$$

### 2.3 General Line Equation (GLE)

#### Properties

1. Generic form of a Linear equation.
2. GLE acts a link between Coordinate Geometry and Linear Algebra for representing a line.

#### Equation #4

General line equation for line a.k.a **2D Hyperplane**:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 = 0
\end{aligned}
$$

#### Equation #5

General line equation for a plane a.k.a **3D Hyperplane**:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_3 x_3 + w_0 = 0
\end{aligned}
$$

#### Equation #6

General line equation for a Hyperplane a.k.a **$n$D Hyperplane**:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + \ldots +  w_n x_n + w_0 = 0
\end{aligned}
$$

#### Why General Line Equation?

A Linear Equation can be represented in either Slope-Intercept Equation (SIE) or General Line Equation (GLE),  
but GLE is preferred over SIE to overcome the two limitations SIE:

1. **Undefined Slope**: GLE simplifies the complexity around undefined slope.
2. **Linear Expansion**: GLE supports higher dimensions beyond 2D Hyper-planes as seen Equation #5 and #6.

> **Note**:
>
> 1. A Line is also called as 2D Hyper-plane.
> 2. A Plane is also called as 3D Hyper-plane.
> 3. A Hyperplane is used in general for greater than 3D.

## 3 ML Context

### 3.1 ML Context

In ML Context:

- $(w_0, w_1, w_2, \ldots, w_n)$ are called as **Weights** or parameters.
- $(x_0, x_1, x_2, \ldots, x_n)$ are **Feature** values.
- $w_0$ is a special type of weight also called as **Bias**.

#### Weights

1. Weights ($w_0, w_1, w_2, \ldots, w_n$) are trainable parameters.
2. $w_0$ is called as **Bias**.
3. Optimization process finds best weights for the line equation that segregates data-points based on their class labels.

#### Features

1. Features $(x_0, x_1, x_2, \ldots, x_n)$ are labeled data-points in the dataset.
2. Label column in the dataset acts as target variable $y$.

#### Bias

1. Intercept $c$ in Coordinate Geometry is called as Bias $w_0$ in Machine Learning.
2. Bias is initialized with some random value at the start of the training.

#### Learning

1. Learning in ML is the process of mathematically identifying best values for weights: $(w_0, w_1, w_2, \ldots, w_n)$ parameters.
2. Decision boundary resulting from this best values of weights combined with features is called as **best fit line**.

### 3.2 Mapping Function

Mapping function for binary classification ML model:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 = 0
\end{aligned}
$$

Mapping function for multi-class classification ML model:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + \ldots +  w_n x_n + w_0 = 0
\end{aligned}
$$

1. $(x_0, x_1, x_2, \ldots, x_n)$ is given to us in the form of dataset.
2. A ML model learns $(w_0, w_1, w_2, \ldots, w_n)$ during learning process.

### 3.3 Half-space

#### Classification Context

In the context of a binary classification problem,

1. The decision boundary separates the data-points into two half-spaces.
2. One side is called as **Positive half-space**.
3. Other side is called as **Negative half-space**.

$$
\large
\begin{aligned}
y = f(x) = w_1 x_1 + w_2 x_2 + w_0
\end{aligned}
$$

##### Positive Half-space

When data-point lies in the positive half-space:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 > 0
\end{aligned}
$$

##### Negative Half-space

When data-point lies in the negative half-space:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 < 0
\end{aligned}
$$

##### On Decision Boundary

When data-point lies on the decision boundary:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 = 0
\end{aligned}
$$
