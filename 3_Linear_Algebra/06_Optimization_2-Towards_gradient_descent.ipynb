## 1 Differentiation

### 1.1 Geometric Representation

#### Slope of a Line

![Slope of a line](images/slope_of_line1.png)

Formula to calculate slope of a line intersecting two points $(x_i, y_1)$ and $(x_2, y_2)$

$$
\large
\begin{aligned}
m &= \frac{y_2 - y_1}{x_2 - x_1} \\[12pt]
  &= \frac{\Delta\;y}{\Delta\;x}
\end{aligned}
$$

Where,

- $\Delta\;x$ is small change in $x$.
- $\Delta\;y$ is small change in $y$.

#### Properties of Slope

- As $x$ increases if $y$ is increasing then slope is positive.
- As $x$ increases if $y$ is decreasing then slope is negative.

#### Slope of Curve

- For a straight line slope is always constant.
- Using above technique we can calculate slope of a line but in real world the Gain/Loss function can be curve (multinomial equation)
- Hence we need to calculate slope of a curve.

#### Measuring Slopes of a Curve

Slope of a curve can be measured in two ways:

1. Using Secant Line
2. Using Tangent Line

### 1.2 Secant

#### Secant Line

If a straight line intersects a curve at two points its called as **Secant**.

#### Slope of Secant Line

![Slope of curve](images/slope_of_secant_line1.png)

#### Formula

Formula to calculate slope of a secant line intersecting a curve at two points $(x_i, y_1)$ and $(x_2, y_2)$

$$
\large
\begin{aligned}
m &= \frac{y_2 - y_1}{x_2 - x_1} \\[12pt]
  &= \frac{f(x_2) - f(x_1)}{x_2 - x_1}
\end{aligned}
$$

### 1.3 Tangent

#### Tangent Line

If a straight line intersects a curve at only one point its called as **Tangent**.

#### Slope of Tangent Line

![Slope of curve derivative](images/slope_of_tangent_line1.png)

##### How to calculate Slope of a Tangent Line?

###### Problem

- For Secant line there are at least two points $(x_1, y_1)$ and $(x_2, y_2)$ to calculate Slope.
- But for Tangent line we have only one point $(x_1, y_1)$ to calculate slope.

###### Solution

- To calculate slope of a Tangent line at $(x_1, y_1)$ a minuscule change in $x_1$ represented as $\Delta{x_1}$ is considered.
- Hence limits are used to measure the rate of change of $x_1$ towards zero.
- Slope of a Tangent line to a curve at a give point is found by calculating the derivative of the function representing that curve.

$$
\large
\begin{aligned}
            m &= \frac{y_2 - y_1}{x_2 - x_1} \\[12pt]
f^{\prime}(x) &= \lim_{\Delta{x}\;\to\;0} \quad \frac{f(x_1 + \Delta\;x) - f(x_1)}{(x_1 + \Delta\;x) - x_1} \\[12pt]
              &= \lim_{\Delta{x}\;\to\;0} \quad \frac{f(x_1 + \Delta\;x) - f(x_1)}{\cancel{x_1} + \Delta\;x - \cancel{x_1}} \\[12pt]
              &= \lim_{\Delta{x}\;\to\;0} \quad \frac{f(x + \Delta\;x) - f(x)}{\Delta\;x}
\end{aligned}
$$

#### Formula

Formula to calculate slope of a tangent line intersecting a curve at only one point $(x_i, y_1)$

$$
\large
\begin{aligned}
f^{\prime}(x) = \frac{d}{dx}f(x) = \lim_{\Delta{x}\;\to\;0} \quad \frac{f(x_1 + \Delta{x}) - f(x_1)}{\Delta{x}}
\end{aligned}
$$

## 2 Derivative

### 2.1 Derivative of a Function

#### Definition

**Q**: What does the derivative of a function represent?

**Ans**:  
The slope of the tangent line to the function's graph at a given point.

#### Why we need Slope?

- $a$ - Some random value of $x$
- $f(a)$ - Value of $f(x)$ at some random value of $x$.
- $f^\prime(a)$ - Slope of $f(x)$:
   - if sign of the slope is **positive** it is in **ascent** (moving towards maxima).
   - if sign of the slope is **negative** it is in **descent** (moving towards minima).

#### How to calculate derivative of a function?

##### Example #1

Compute $f^\prime(x^2)$

$$
\large
\begin{aligned}
\frac{d}{dx} x^2 &= \lim_{\Delta{x}\;\to\;0} \quad \frac{(x + \Delta{x})^2 - x^2}{\Delta\;x} \\[12pt]
                 &= \lim_{\Delta{x}\;\to\;0} \quad \frac{\cancel{x^2} + {\Delta{x}}^2 + (2\cdot{x}\cdot\Delta{x}) - \cancel{x^2}}{\Delta\;x} \\[12pt]
                 &= \lim_{\Delta{x}\;\to\;0} \quad \frac{\cancel{\Delta{x}}\cdot(\Delta{x} + 2\cdot{x})}{\cancel{\Delta\;x}} \\[12pt]
                 &= \lim_{\Delta{x}\;\to\;0} \quad \Delta{x} + (2\cdot{x}) \\[12pt]
                 &= \qquad \qquad 0 + (2\cdot{x}) \\[12pt]
                 &= \qquad \qquad 2\cdot{x}
\end{aligned}
$$

Derivative of $f(x) = x^2$: https://www.desmos.com/calculator/uuq8hnbmis

#### First Derivative

- Taking derivative of a function $f(x)$ once is called **First Derivative**.
- Taking the derivative of First Derivative of a function is called **Second Derivative**.

### 2.2  Derivability of a Function

#### Criteria

A function $f(x)$ is considered Derivable if **Left Derivative = Right Derivative**.

**Q**: What is a crucial step in determining if a function is differentiable?

**Ans**:  
Verifying if the first derivative, is continuous or discontinuous.

#### Non Derivable Function

##### Example #1

Derivability of function $f(x) = \vert{x}\vert$

$$
\large
\begin{aligned}
\frac{d}{dx}\vert{x}\vert &= \frac{d}{dx}({-x}) = -1 \\[5pt]
\frac{d}{dx}\vert{x}\vert &= \frac{d}{dx}(x)    =  1 \\[5pt]
\end{aligned}
$$

$$
\large
\begin{aligned}
\lim_{x\;\to\;0^-}\;\frac{d}{dx}\vert{x}\vert = -1 &                 & \lim_{x\;\to\;0^+}\;\frac{d}{dx}\vert{x}\vert = 1 \\[5pt]
                                                   & \qquad -1 \ne 1 &
\end{aligned}
$$

$$
\large
\begin{aligned}
\text{Left Derivative} \ne \text{Right Derivative} \\[5pt]
\text{The derivative is discontinuous at}\;x = 0
\end{aligned}
$$

> **Note**:  
> 
> **Every differentiable** function is **continuous** but **every continuous** function is **not differentiable**.

### 2.4 Rules of Differentiation

#### 1 Constant Rule

$$
\large
\begin{aligned}
\frac{d}{dx} \Bigl[c \cdot f(x) \Bigr] = c \cdot \frac{d}{dx}f(x)
\end{aligned}
$$

#### 2 Sum Rule

$$
\large
\begin{aligned}
\frac{d}{dx} \Bigl[f(x) + g(x)\Bigr] = \frac{d}{dx}f(x) + \frac{d}{dx}g(x)
\end{aligned}
$$

#### 3 Difference Rule

$$
\large
\begin{aligned}
\frac{d}{dx} \Bigl[f(x) - g(x)\Bigr] = \frac{d}{dx}f(x) - \frac{d}{dx}g(x)
\end{aligned}
$$

#### 4 Product Rule

$$
\large
\begin{aligned}
\frac{d}{dx} \Bigl[f(x) \cdot g(x)\Bigr] = g(x)\frac{d}{dx}f(x) + f(x)\frac{d}{dx}g(x)
\end{aligned}
$$

#### 5 Quotient Rule

$$
\large
\begin{aligned}
\frac{d}{dx} \Biggl[\frac{f(x)}{g(x)}\Biggr] = \frac{g(x)\frac{d}{dx}f(x) - f(x)\frac{d}{dx}g(x)}{\bigl[g(x)\bigr]^2}
\end{aligned}
$$

#### 6 Chain Rule

$$
\large
\begin{aligned}
\frac{d}{dx} f\Bigl(g(x)\Bigr) = \frac{d}{dx}f\Bigl(g(x)\Bigr) \cdot \frac{d}{dx}g(x)
\end{aligned}
$$

##### How Chain rule works?

- Keep the inside, take derivative of outside.
- Multiply it by the derivative of the inside.

##### Example #1

$$
\require{cancel}
\large
\begin{aligned}
\frac{d}{dx} \log(x^2) &= \frac{1}{x^2} \cdot 2x  \\[5pt]
                       &= \frac{1}{x\cancel{x}} \cdot 2\cancel{x} \\[5pt]
                       &= \frac{2}{x}
\end{aligned}
$$

- Keep the inside $x^2$ as it is and take derivative of outside $\log$ i.e., $\frac{1}{x^2}$.</li>
- Multiply it by the derivative of the inside $x^2$ i.e., $2x$.

##### Example #2

$$
\large
\begin{aligned}
\frac{d}{dx} e^{-x} &= e^{-x} \cdot (-1) \\[5pt]
                    &= -{e^{-x}}
\end{aligned}
$$

### 2.5 Critical Points

![Critical Points](images/critical_points1.png)

#### Definition

Slope of the Tangent line to a curve (i.e., derivative of a function) will be **zero at Critical Points**.

$$
\large
\begin{aligned}
f^{\prime}(x) = 0
\end{aligned}
$$

#### Properties

- Critical points can be maxima, minima or saddle point.

### 2.6 Summary

- Derivative is a single variable function.
- Calculating first derivative of a function simplifies the function.
- Slope is always constant for straight line, but it keeps changing (increasing/decreasing) at every point for a complex $f(x)$

## 3 Double Derivative

### 3.2 Local Optima

#### Why calculate Double Derivative?

Calculating double derivative of a function helps us in identifying **Local Optima**.

#### What is Local Otima?

A local optimum is a point where the function is better than all nearby points.

Types of Local Optima:

1. Maxima
2. Minima

##### Maxima

A local maximum is a critical point on a function's graph where the value of  
$f(x)$ is higher than all nearby points in its immediate vicinity.

$$
\large
\begin{aligned}
f^{\prime\prime}(x) < 0 \quad \text{Local Maxima}
\end{aligned}
$$

##### Minima

A local minimum is a critical point on a function's graph where the value of  
$f(x)$ is lower than or equal to all nearby points in its immediate vicinity.

$$
\large
\begin{aligned}
f^{\prime\prime}(x) > 0 \quad \text{Local Minima}
\end{aligned}
$$

#### Saddle Point

Non-optimal points where both $f^{\prime}(x) = 0$ (First derivative) and $f^{\prime\prime}(x) = 0$ (Second derivative) are zero.

$$
\large
\begin{aligned}
f^{\prime\prime}(x) = 0 \quad \text{Saddle point}
\end{aligned}
$$

#### Why calculate Local Optima?

A **Local Optimum** is a solution to an optimization problem that is the best (maximum or minimum) within  
a small neighborhood of nearby solutions, but not necessarily the best overall (global optimum).

### 3.3 Summary

1. Using derivatives, identify **Critical points** (where slope is zero).
2. Critical points can be maxima or minima.
3. Using double derivatives, identify maxima or minima.
   - If double derivative is **greater than zero** then the critical point is **Minima**.
   - If double derivative is **less than zero** then the critical point is **Maxima**.
   - If double derivative is **equal to zero** then the critical point is **Saddle point**.

$$
\large
\begin{aligned}
f^{\prime\prime}(x)
\begin{cases} 
f^{\prime\prime}(x) > 0 \quad \text{Local Minima} \\[15pt]
f^{\prime\prime}(x) < 0 \quad \text{Local Maxima} \\[15pt]
f^{\prime\prime}(x) = 0 \quad \text{Saddle point} 
\end{cases}
\end{aligned}
$$

### 3.4 General Framework

#### First and Second Derivative Test for Extrema

##### STEP 1: Compute First Derivative

Given function $f(x)$ compute its first derivative:

$$
\large
\begin{aligned}
f^\prime(x)
\end{aligned}
$$

##### STEP 2: Find Critical Point

Equate the first derivative to zero to find the critical point:

$$
\large
\begin{aligned}
f^\prime(x) = 0
\end{aligned}
$$

Critical Point :- Critical value of $x$, say $c$, at which $f^\prime(x) = 0$.

##### STEP 3: Classify Using Second Derivative

To classify the Critical point from STEP 2 to the correct type of local Optima, compute derivative of $f^\prime(x)$ i.e., second derivative of $f(x)$ from STEP 1:

$$
\large
\begin{aligned}
f^{\prime\prime}(x)
\end{aligned}
$$

Identify the type of Local Optima based on below rule:

$$
\large
\begin{aligned}
f^{\prime\prime}(x)
\begin{cases} 
\text{if} \quad f^{\prime\prime}(x) > 0: \quad \text{Local Minima} \\[8pt]
\text{if} \quad f^{\prime\prime}(x) < 0: \quad \text{Local Maxima} \\[8pt]
\text{if} \quad f^{\prime\prime}(x) = 0: \quad \text{Test Inconclusive}
\end{cases}
\end{aligned}
$$

##### STEP 4: Calculate Local Optima Value

Substitute the Critical Value $c$ from STEP 2 in original function $f(x)$ to identify the actual value of Local Optima.

$$
\large
\begin{aligned}
f(c) \in \mathbb{R}
\end{aligned}
$$
