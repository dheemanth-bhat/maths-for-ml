## 1 Shifting of Lines

### 1.1 One Direction

Shifting lines in one direction (x-axis or y-axis) by $a$ units:

$$
\large
\begin{array}{r}
\text{x-axis} \mid & \text{Right} \rightarrow & w_1(x_1 - a) & + &      w_2 x_2 & + & w_0 = 0 \\[5pt]
\text{x-axis} \mid &  \text{Left} \leftarrow  & w_1(x_1 + a) & + &      w_2 x_2 & + & w_0 = 0 \\[5pt]
\text{y-axis} \mid &    \text{Up} \uparrow    &      w_1 x_1 & + & w_2(x_2 - a) & + & w_0 = 0 \\[5pt]
\text{y-axis} \mid &  \text{Down} \downarrow  &      w_1 x_1 & + & w_2(x_2 + a) & + & w_0 = 0 \\[5pt]
\end{array}
$$

### 1.2 Two Directions

Shifting lines in two direction (x-axis and y-axis) simultaneously by $a$ units:

$$
\large
\begin{aligned}
w_1(x_1 - a) + w_2(x_2 - a) + w_0 &= 0 \\[5pt]
w_1 x_1 - w_1 a + w_2 x_2 - w_2 a + w_0 &= 0 \\[5pt]
w_1 x_1 + w_2 x_2 - w_1 a - w_2 a + w_0 &= 0 \\[5pt]
w_1 x_1 + w_2 x_2 - a(w_1 + w_2) + w_0 &= 0 \\[5pt]
w_1 x_1 + w_2 x_2 + w_0^\prime &= 0
\end{aligned}
$$

Where $w_0^\prime$ is:

$$
\large
\begin{aligned}
w_0^\prime = w_0 - a(w1 + w2)
\end{aligned}
$$

#### Properties

- Weights vector $w$ control orientation (slope).
- Bias $w_0$ controls shifting (translation).

## 2 Distances

The probability that a data point belongs to the correct class is directly proportional to the distance of a point from the decision boundary.

1. Points closer to the Decision Boundary have lower probability of being in their class.
2. Points closer to the Decision Boundary have higher probability of being in their class.
3. Points lying on the Decision Boundary have 0.5 probability of belonging to either class.

### 2.1 Distance of Vector from Origin

#### Formula

Formula to calculate distance of a vector $\vec{w}$ from origin:

$$
\large
\begin{aligned}
d = \frac{-w_0}{\Vert{w}\Vert}
\end{aligned}
$$

#### Properties

The minus sign in the formula shows the direction. When the distance is:

1. Negative: Decision boundary and the vector are in the **opposite direction** from origin.
2. Positive: Decision boundary and the vector are in the **same direction** from origin.

### 2.2 Distance of Point from Line

#### Formula

Formula to calculate the distance between a point $x_0$ and line (Decision Boundary) using vector $\vec{w}$:

$$
\large
\begin{aligned}
d = \frac{w^{T}x_0 + w_0}{\Vert{w}\Vert}
\end{aligned}
$$

#### Properties

When the distance is:

1. Greater than zero, data-points are in the same direction of Normal vector $\vec{w}$ and are considered Positive (+1).
2. Less than zero, data-points are in the opposite direction of Normal vector $\vec{w}$ and are considered Negative (-1).
3. Equal to Zero, data-points lies on the Normal vector $\vec{w}$ and be considered Positive or Negative depending on the usecase.

#### Examples

##### Code

```python
import numpy as np
from fractions import Fraction as F
```

```python
def distance_bw_point_and_line(x, w, w_0):
    """
    Function to calculate distance between data-point `x`
    and the Decision Boundary using its Normal vector `w`.
    """
    dot_p = np.dot(np.asarray(w), np.asarray(x))
    # print("Dot product:", dot_p)

    # L2 Norm of `w` vector.
    mag = np.linalg.norm(w)
    # print("Magnitude:", round(mag, 6))

    dist = (dot_p + w_0) / mag
    return round(dist, 6)
```

Find the distance of the point (3, 9) from the plane $-2x + 3y -3 = 0$?

```python
d1 = distance_bw_point_and_line(x=[3, 9], w=[-2, 3], w_0=-3)
print("Distance:", d1)
```

### 2.3 Distance Between Planes

#### Definition

To calculate distance between two parallel planes:

1. When planes have same magnitude, divide the difference of their constants by their normal vector's magnitude.
2. If two vectors $w^{(1)}$ and $w^{(2)}$ are of different magnitudes, then they must satisfy the condition: $w^{(1)} = \lambda \cdot w^{(2)}$

#### Formula

Formula to calculate distance between two planes with same magnitude:

$$
\large
\begin{aligned}
d = \frac{w_0^{(1)} \;-\; w_0^{(2)}}{\Vert{w}\Vert}
\end{aligned}
$$

Formula to calculate distance between two planes of different magnitude where $w^{(1)} = \lambda \cdot w^{(2)}$:

$$
\large
\begin{aligned}
d = \frac{w_0^{(1)}}{\Vert{w^{(1)}}\Vert} \;-\; \frac{w_0^{(2)}}{\Vert{w^{(2)}}\Vert}
\end{aligned}
$$

#### Properties

1. Given two parallel lines can have same weight vectors **except Intercept** $w_0$.


#### Examples

##### Code

```python
def distance_bw_parallel_lines(w_01, w_02, w1, w2=None):
    """
    Function to calculate distance between two parallel lines
    using their Normal vectors `w1` and `w2` and constants.
    """
    if w2 is None:
        w2 = w1
    else:
        w1_uv = np.asarray(w1) / np.linalg.norm(w1)
        w2_uv = np.asarray(w2) / np.linalg.norm(w2)

        if not np.array_equal(w1_uv, w2_uv):
            # Check if the unit normals of `w1` and `w2` vectors are identical.
            raise ValueError("Unit normals of `w1` and `w2` are not identical.")

    # L2 Norm of `w1` vector.
    mag1 = np.linalg.norm(w1).item()

    # L2 Norm of `w2` vector.
    mag2 = np.linalg.norm(w2).item()

    # Calculate distance between lines.
    dist = (w_01 / mag1) - (w_02 / mag2)

    return round(dist, 6)
```

##### Example #1

Calculate distance between:

1. Line #1: $w^T{x} + 5 = 0$
2. Line #2: $w^T{x} + 3 = 0$

where $w = [3, 2, 1]$.

```python
d1 = distance_bw_parallel_lines(w_01=5, w_02=3, w1=[3, 2, 1])
print("Distance:", d1)
```

## 3 Half-spaces

### 3.1 Binary Classification

#### Types of Half-Spaces

In Binary Classification there are two types of Half-spaces:

1. Positive Half-space
2. Negative Half-space

#### Best Decision Boundary

There are two ways to identify the best line (Decision Boundary) that segregates the positive and negative half-spaces:

1. Using Direction/Angle
2. Using Distance/Magnitude

##### 1 Using Direction

1. Weight vector will always be perpendicular to the line (Decision Boundary).
2. All the data-points that lies in the opposite direction of the weight vector will be Negative i.e., $\cos \uptheta$ between (90, 270).

##### 2 Using Distance

When the distance is:

1. Greater than zero, data-points are in the Positive (+1) Half-space.
2. Less than zero, data-points are in the Negative (-1) Half-space.
3. Equal to Zero, data-points lies on the Normal vector and be considered Positive or Negative class depending on the usecase.

#### Mapping Function

##### 1 Positive Half-space

When data-point lies in the positive half-space:

$$
\large
\begin{aligned}
w^T x + w_0 > 0
\end{aligned}
$$

##### 2 Negative Half-space

When data-point lies in the negative half-space:

$$
\large
\begin{aligned}
w^T x + w_0 < 0
\end{aligned}
$$

##### On Decision Boundary

When data-point lies on the decision boundary:

$$
\large
\begin{aligned}
w^T x + w_0 = 0
\end{aligned}
$$
