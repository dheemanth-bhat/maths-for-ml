## Decision Boundary

### 1.1 Best Decision Boundary

A Line can be considered as best Decision Boundary if the **average distance** of all data-points from either side of the half-spaces is **maximum** compared to all other possible lines.

#### Properties

1. Farther the distance of the data-points from the line on either side of the half-space, better the separator (Decision Boundary).
2. Positive and negative **distances cancel out each other** if they are directly summed up.
3. There are three tricks to resolve the problem of canceling distances.
   1. Absolute values of distances $\vert{d}\vert$
   2. Square the distances
   3. Multiply class label $y_i$ to each distance $d_i$
4. Trick #3 is used in Optimization function since it correctly penalizes the misclassification compared to other two tricks.

### 1.2 Gain Function

#### Formula

$$
\large
\begin{aligned}
g(x, y, \vec{w}, w_0) = \arg\max_{w, w_0} \frac{1}{n} \sum_{i=1}^n \frac{w^T x^{(i)} \;+\; w_0}{\Vert{w}\Vert} \cdot y^{(i)}
\end{aligned}
$$

**Maximize Gain function.**

### 1.3 Loss Function

#### Definition

$$
\large
\begin{aligned}
\text{Loss Function} = -\;\text{Gain Function}
\end{aligned}
$$

#### Formula

$$
\large
\begin{aligned}
l(x, y, \vec{w}, w_0) = \arg\min_{w, w_0} \frac{1}{n} \sum_{i=1}^n \frac{w^T x^{(i)} \;+\; w_0}{\Vert{w}\Vert} \cdot y^{(i)}
\end{aligned}
$$

**Minimize Loss function.**

#### Properties

1. Loss function is also considered as **Error**.
2. Percentage of miss-classified observations can also be considered as Error.
