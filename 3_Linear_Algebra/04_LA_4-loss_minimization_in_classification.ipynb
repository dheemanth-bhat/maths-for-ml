# Linear Algebra 4 - Loss minimization in classification

## 1 Decision Boundary

### 1.1 Half-Space

#### Positive Half-space

In Positive Half-space all data-points have **distance greater than zero** because,

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 > 0
\end{aligned}
$$

#### Negative Half-space

In Negative Half-space all data-points have **distance less than zero** because,

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 < 0
\end{aligned}
$$

#### Decision Boundary

On Decision Boundary all data-points have **distance equal to zero** because,

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 = 0
\end{aligned}
$$

### 1.2 Best Decision Boundary

#### Objective

- While identifying a best fit line, a Line can be considered as best Decision Boundary if the **average distance** of all data-points from either side of the half-spaces is **maximum** compared to all other possible lines.
- Penalizing miss-classification.

#### Properties

1. Farther the distance of the data-points from the line on either side of the half-space, better the separator (Decision Boundary).
2. Positive and negative **distances cancel out each other** if they are directly summed up.
3. There are three tricks to resolve the problem of canceling distances.
   1. Absolute values of distances $\vert{d}\vert$
   2. Square the distances
   3. Multiply class label $y_i$ to each distance $d_i$
4. **Absolute** function $y = f(x) = \vert{x}\vert$ is **not differentiable** hence Trick #1 is not used.
5. Even though **Square** operation is differentiable, it is **computationally expensive** compared to Trick #3.
6. Trick #3 is used in Optimization function since it correctly penalizes the misclassification compared to other two tricks.

### 1.3 Mathematical Notation

Mathematical notations for a Binary Classification problem.

#### Dataset

$$
\large
\begin{aligned}
D = \biggl\{ \bigl(x_i, y_i \bigr) \biggr\}_{i=1}^n
\end{aligned}
$$

Where,

- $D$ is Dataset
- $x$ are features
- $y$ are labels
- $n$ is the size of the dataset

#### Label

Actual Label $y$ can take values:

$$
\large
\begin{aligned}
y_i 
\begin{cases} 
+1 \quad \text{in +ve Half-Space} \\[15pt]
-1 \quad \text{in -ve Half-Space} 
\end{cases}
\end{aligned}
$$

#### Classifier

Equation of Decision Boundary:

$$
\large
\begin{aligned}
w_1 x_1 + w_2 x_2 + w_0 &= 0 \\[3pt]
              \text{or} &    \\[3pt]
            w^T x + w_0 &= 0
\end{aligned}
$$

Predicted value $\hat{y_i}$ can take values:

$$
\large
\begin{aligned}
\hat{y_i}
\begin{cases} 
+1 \quad \text{if} \; \bigl(w^T x + w_0\bigr) \ge 0  \\[15pt]
-1 \quad \text{if} \; \bigl(w^T x + w_0\bigr) < 0 
\end{cases}
\end{aligned}
$$

### 1.4 Gain Function

#### Formula

Equation of Gain Function for Binary Classification:

$$
\large
\begin{aligned}
g(D, \vec{w}, w_0) = \frac{1}{n} \sum_{i=1}^n \frac{w^T x^{(i)} \;+\; w_0}{\Vert{w}\Vert} \cdot y^{(i)}
\end{aligned}
$$

#### Objective

**Maximize Gain function**.

### 1.5 Loss Function

#### Definition

$$
\large
\begin{aligned}
\text{Loss Function} = -\;\text{Gain Function}
\end{aligned}
$$

#### Formula

Equation of Loss Function for Binary Classification:

$$
\large
\begin{aligned}
l(D, \vec{w}, w_0) = \frac{1}{n} \sum_{i=1}^n \frac{w^T x^{(i)} \;+\; w_0}{\Vert{w}\Vert} \cdot y^{(i)}
\end{aligned}
$$

#### Objective

**Minimize Loss function**.

#### Properties

1. Loss function is also considered as **Error**.
2. Percentage of miss-classified observations can also be considered as Error.

## 2 Perceptron Learning

### 2.1 Handling Miss-Classification

#### What is Miss-classification?

when the actual label $y_i$ and the predicted label $\hat{y}$ are not same the data-point is miss-classified.

#### How to fix miss-classification?

1. If a **label is positive** but miss-classified in **Negative half-space**, rotate the line in **clockwise** direction.
2. If a **label is negative** but miss-classified in **Positive half-space**, rotate the line in **anticlockwise** direction.

### 2.2 Perceptron Learning Algorithm (PLA)

PLA's weight updation logic:

$$
\large
\begin{array}{l}
\text{if sign}\bigl(w^T x_i + w_0 \bigr) \ne \text{sign}\bigl(y_i\bigr): \\[5pt]
\qquad w = w + x_i \cdot y_i \\[3pt]
\qquad w_0 = w_0 + y_i
\end{array}
$$
