## 2 Measure of Central Tendency

Measuring the centrality of data.  

Why do we have to measure the centrality of data?  

When we measure the centrality of data, we are essentially trying to summarize a dataset with a single representative value.  
Real-world datasets can have millions of points. Instead of listing all of them, we want a compact representation.

#### Types of Central tendencies

1. Mean
2. Median
3. Mode

### 2.1 Mean

#### Definition

Sum of all values in the array divided by number of elements in the array.

#### Formula

Given an array of numbers $x = \{x_1, x_2, x_3, \ldots, x_n\}$ of length $n$. Mean can be calculated as:


$$
\large
\begin{aligned}
\mu = \frac{1}{n} \sum_{i=1}^{n} x_i
\end{aligned}
$$

Mean is represented using Greek letter $\mu$.

#### Issues with Mean

1. Mean is affected by outliers.
2. Equal weightage problem i.e., mean gives identical importance (weight) to every element in the array.

### 2.2 Median

#### Definition

1. Midpoint of **sorted** data.
2. Median is **robust** (not affected) to outliers.

#### Formula

Given an array of numbers $x = \{x_1, x_2, x_3, \ldots, x_n\}$ of length $n$:

##### Case #1: Odd number of elements in array $x$

Median is element at index $\displaystyle \frac{n + 1}{2}$:

$$
\huge
\begin{aligned}
x_{\frac{n + 1}{2}}
\end{aligned}
$$

##### Case #2: Even number of elements in array $x$

Median is mean of elements at index $\displaystyle \frac{n}{2}$ and $\displaystyle \frac{n}{2} + 1$:

$$
\huge
\begin{aligned}
\frac{1}{2}\bigl(x_{(\frac{n}{2})}\;+\;x_{(\frac{n}{2} + 1)}\bigr)
\end{aligned}
$$

or

$$
\huge
\begin{aligned}
\frac{x_{(\frac{n}{2})}\;+\;x_{(\frac{n}{2} + 1)}}{2}
\end{aligned}
$$

### 2.3 Mode

#### Definition

Mode gives the most frequent elements.  

Dataset can have more than one mode. There are following types of modes:

1. Uni-Model data
2. Bi-Model data
3. Multi-Model data

> **Note**:
> 
> Mean and Median are suitable for continuous data. Mode is suitable for discrete data.

### 2.4 Weighed Mean

#### Definition

A weighted average is a way to find the average of a set of numbers when some numbers are more important than others, giving them more "weight" in the final calculation.  
It is calculated by multiplying each data point by its assigned weight, summing these products, and then dividing by the sum of all the weights.

#### Formula

$$
\large
\begin{aligned}
\bar{x} = \frac{\sum_{i = 1}^{n} w_i * x_i}{\sum_{i = 1}^{n} w_i}
\end{aligned}
$$

> **Note**:
>
> Weighed mean is used while calculating Loss function for imbalanced dataset.

## 3 Measure of Variability

#### Variability

Variability is spread of data.

#### Types of Variability

1. Range
2. Variance
3. Standard Deviation
4. Inter-quartile range (IQR)

### 3.1 Range

#### Definition

- Range measures the spread in data between the element with largest value and smallest value.
- Distance between smallest observation to the largest observation on the number line.

#### Formula

$$
\large
\begin{aligned}
\text{Range} = \text{Max element} - \text{Min element}
\end{aligned}
$$

#### Properties

- Smaller the Range, smaller is the spread or larger the Range, larger is the spread in data. 
- Range is not affected by the sign (+/-) of the element.
- Range is sensitive to outliers. Instead of range, a variation on range called as Inter Quartile Range is used.

### 3.2 Quantiles

#### Definition

A quantile is a value that divides a probability distribution (or dataset) into intervals with equal probabilities.  
In other words, if you sort your data in increasing order, quantiles are the "cut points" that split the data into equal-sized groups.

#### Properties

1. Quantiles break ordered/sorted data into equal parts.
2. Some popular Quantiles are Percentiles, Quartiles and Deciles.
3. **Quartiles** are a special type of Quantiles that divide ordered data into **4 equal parts**.
4. **Percentiles** are a type of Quantiles that divide ordered data into **100 equal parts**.

### 3.3 Inter-Quartile Range (IQR)

![IQR](images/IQR.png)

#### Definition

1. IQR is difference between value of element at 75th percentile and value of element at 25th percentile.
2. IQR measures the statistical dispersion i.e., it describes the spread in the data.
3. IQR is median-based measure of spread, i.e., it tells us how far the data-points are from the Median.

##### Cut points

* Q1 is cut point at 25th percentile
* Q2 is cut point at 50th percentile (aka median)
* Q3 is cut point at 75th percentile

#### Formula

$$
\large
\begin{aligned}
\text{IQR} = \text{Q3} - \text{Q1}
\end{aligned}
$$

#### Properties

1. IQR holds the middle 50% of the data.
2. Larger the IQR higher the spread in data aka variance.
3. IQR uses **Median as central tendency** (center of data).

#### Example

##### Example #1

```python
x = np.array([-30, 8, 5, 1, 9, 0, 2, 7, 4, 3, 50, 6])
x
```

```python
plt.figure(figsize=(8, 2))
sns.boxplot(x=x)
plt.show()
```

```python
# Just for visualization.
np.sort(x)  # Optional
```

```python
q1 = np.percentile(x, 25)
print("25th percentile:", q1)
```

```python
q2 = np.percentile(x, 50)
print("50th percentile (median):", q2)
```

```python
q3 = np.percentile(x, 75)
print("75th percentile:", q3)
```

```python
iqr = q3 - q1
print("IQR:", iqr)
```

### 3.4 Variance

#### Definition

- Variance is a statistical measure that quantifies how spread out a set of data points are around their average (mean).
- Average of the square of the distance of each element from mean.

#### Formula

$$
\large
\begin{aligned}
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
\end{aligned}
$$

#### Properties

1. Variance measures the **statistical dispersion** i.e., it describes the spread in the data.
2. Variance is mean-based measure of spread i.e., it tell us how far the data-points are from the Mean.
3. A low variance indicates data is clustered closely to the mean, while a high variance means data is widely scattered.
4. In most cases, lesser the Variance in dataset its better.
5. Variance uses **Mean as central tendency** hence it is highly sensitive to outliers.

### 3.5 Standard Deviation

#### Definition

- Standard deviation is a statistical measure (normalized) that quantifies how spread out the data points are around their average.
- Standard Deviation is square root of Variance.

#### Formula

$$
\large
\begin{aligned}
\text{Standard Deviation} &= \sqrt{\text{Variance}} \\[10pt]
                   \sigma &= \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
\end{aligned}
$$

Standard Deviation is represented using Greek letter $\sigma$.

#### Properties

1. Standard Deviation is Normalizing the variance to the proper scale of data.
2. Standard Deviation represents how far are the data-points are from the mean.
3. In most cases, lesser the Standard Deviation in dataset its better.
